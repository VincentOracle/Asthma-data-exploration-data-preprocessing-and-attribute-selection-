```python
## import necessary modules
import pandas as pd

aggregation = pd.read_csv(
    filepath_or_buffer= "data/aggregated.csv", # For Windows, use "\\" instead of "/"
    sep = ",", # comma separated file
    header = 0, # field names in the first line. Index starts with 0 in Python
    encoding = "utf-8", # encoding 
    #parse_dates = [1,33] # If there is any date field, put index (indices) of the field(s)
)

## Set the maximum number of columns as "None" to display all columns
pd.set_option('display.max_columns', None)

## print first 5 lines. If you need more lines, put the number of lines as an argument
## of the head function
print(aggregation.head(10)) 
print()
print(aggregation.shape) ## show the number of rows and columns
```

```python
aggregation[["ASPRIN53"]].describe()
```

```python
aggregation["ASPRIN53"].value_counts()
```

```python
print(aggregation["ASPRIN53"] == 1)
```

```python
# The next is to create subsets of samples according to taking aspirin or not
taking_aspirin = aggregation[ aggregation["ASPRIN53"] == 1]
taking_aspirin.head()
```

```python
not_taking_aspirin =  aggregation[ aggregation["ASPRIN53"] == 2]
not_taking_aspirin.head()
```

```python
## Sample 100 instances for each case

sample_taking_aspirin = taking_aspirin.sample(n=100, random_state=1)
sample_taking_aspirin.head()
print(sample_taking_aspirin.shape)
```

```python
sample_not_taking_aspirin = not_taking_aspirin.sample(n=100, random_state=1)
sample_not_taking_aspirin.head()
print(sample_not_taking_aspirin.shape)
```

```python
sample_with_aspirin = list(sample_taking_aspirin["TOTEXP09"])
sample_with_aspirin
```

```python
sample_without_aspirin = list(sample_not_taking_aspirin["TOTEXP09"])
sample_without_aspirin
```

```python
expenditure = []
expenditure.extend(sample_with_aspirin)
expenditure.extend(sample_without_aspirin)

aspirin = []

for num in range(200):
    if num < 100:
        aspirin.append("Taking Aspirin")
    else:
        aspirin.append("Not Taking Aspirin")
    
```

```python
d = {'Aspirin': aspirin,'Expenditure':expenditure}

df = pd.DataFrame(d)

df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(rc={'figure.figsize':(10 ,12)})

sns.boxplot( x=df["Aspirin"], y=df["Expenditure"]);
plt.show()
```
